```python
# Generate csv dataset from croped hand gesture images 
import re
import os
import numpy as np

#add csv to write mediapipe coordinates to a csv file
import csv
import cv2

# For static images:
#IMAGE_FILES = ['data/che475.jpg']

def normalize(un_list):
#     norm_list = [(float(i)-min(un_list))/(max(un_list)-min(un_list)) for i in un_list]
    norm = un_list/255
    return norm

#generate row (1 csv row for each image)
def generate_row(path, label):
    # Read an image
    img = cv2.imread(path)
    # Convert to grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Flatten img
    data = np.array(img_gray)
    flattened = data.flatten()
    
#     print(img_gray.shape)
#     print(flattened.shape)
    
    #normalize pixel values between 0 & 1
    normalized = normalize(flattened)
    
    row = [label]
#     head = ['label']
           
    for inx,pxl in enumerate(normalized):
#         head.append('pixel'+ str(inx))
        row.append(pxl)
    
#     print (row)
    with open("esl_data.csv","a") as fp:
        wr = csv.writer(fp,dialect='excel')
#         wr.writerow(head)
        wr.writerow(row)
```

```python
#transverse directory and generate landmarks for all images in the directory
import os
cwd = os.getcwd()
path = os.path.join(cwd, "data2")
file_list = os.listdir(path)
#print(plist)
label = []
filename = []

for fname in file_list :
    #print(fname)
    #Remove file extention
    base = os.path.splitext(fname)[0]
    #remove numbers from file name and use it as a label
    l = re.sub('[0-9]','',base)
    label.append(l)
    
    path="data2/"+ fname
    filename.append(path)
#     print(path)
#     print("Label: {} , Filename: {}".format(l,fname))
    #generate excel row for each image in data folder
    generate_row(path, l)
```

```python
#clean csv - remove empty rows
def clean_csv(path, new_path):
    with open(path) as input , open (new_path, 'w', newline='') as output:
        writer = csv.writer(output)
        for row in csv.reader(input):
            if any(field.strip() for field in row):
                writer.writerow(row)
                
clean_csv('esl_data.csv','esl_data_clean.csv')
```
