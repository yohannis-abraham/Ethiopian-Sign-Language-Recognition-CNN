## CNN with MediaPipe for Ethiopian Sign Language Recognition

Here, we use **OpenCV** and **MediaPipe** to generate a hand landmark dataset from our dataset image directory. We normalize values and clean the CSV file by removing empty spaces. Then, we use the prepared dataset to train our CNN model and evaluate the model accordingly.

### Generate a dataset from hand landmarks

```python
# Generate landmark dataset from images 
import re
import os

#add csv to write mediapipe coordinates to a csv file
import csv
import cv2
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands

# For static images:
#IMAGE_FILES = ['data/che475.jpg']

def normalize(un_list):
    norm_list = [(float(i)-min(un_list))/(max(un_list)-min(un_list)) for i in un_list]
    return norm_list

def generate_landmarks(path, label):
    with mp_hands.Hands(
    static_image_mode=True,
    max_num_hands=2,
    min_detection_confidence=0.5) as hands:
        # Read an image, flip it around y-axis for correct handedness output (see # above).
        image = cv2.flip(cv2.imread(path), 1)
        x , y, c = image.shape
        #print(image.shape)
    
        # Convert the BGR image to RGB before processing.
        results = hands.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
            
        if results.multi_hand_landmarks:
            image_height, image_width, _ = image.shape
            annotated_image = image.copy()
            landmarks = []
            for hand_landmarks in results.multi_hand_landmarks:
                for lm in hand_landmarks.landmark:
            # print(id, lm)
                    lmx = int(lm.x * x)
                    lmy = int(lm.y * y)

                    landmarks.append(lmx)
                    landmarks.append(lmy)
                mp_drawing.draw_landmarks(
                annotated_image,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS,
                mp_drawing_styles.get_default_hand_landmarks_style(),
                mp_drawing_styles.get_default_hand_connections_style())
            #cv2.imwrite('annotated_image' + str(idx) + '.png', cv2.flip(annotated_image, 1))
            
            #normalize landmark values
            landmark_normalized = normalize(landmarks)
#            print('hand_landmarks:', landmarks)
            row = [label]
           
            for land in landmark_normalized:
                row.append(land)
            print (row)
            with open("esl_data_mediapipe.csv","a") as fp:
                wr = csv.writer(fp,dialect='excel')
                wr.writerow(row)

#transverse directory and generate landmarks for all images in the directory
import os
cwd = os.getcwd()
path = os.path.join(cwd, "data_rgb")
file_list = os.listdir(path)
#print(plist)
label = []
filename = []

for fname in file_list :
    #print(fname)
    #Remove file extention
    #filename.append(fname)
    base = os.path.splitext(fname)[0]
    #remove numbers from file name and use it as a label
    l = re.sub('[0-9]','',base)
    label.append(l)
    
    path="data_rgb/"+ fname
    filename.append(path)
    #print(path)
    #print("Label: {} , Filename: {}".format(l,fname))
    #generate landmark for image in the dataset
    generate_landmarks(path,l)

```

```python
#clean csv - remove empty rows
def clean_csv(path, new_path):
    with open(path) as input , open (new_path, 'w', newline='') as output:
        writer = csv.writer(output)
        for row in csv.reader(input):
            if any(field.strip() for field in row):
                writer.writerow(row)
                
clean_csv('esl_data_mediapipe.csv','esl_data_mediapipe_clean.csv')
```

### Load dataset and pre-process it for training 

```python
#imports
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Load the train and test data
train = pd.read_csv('esl_data_mediapipe_clean.csv')
#test = pd.read_csv('sign_mnist_test.csv')
# ectract lables
labels = train['label'].values
unique_value = np.array(labels)
np.unique(unique_value)

dlist = ["label"]
cols =[x for x in train.columns.values.tolist() if x not in dlist]

print(train['x1'])
print(train.values.shape)
#drop lable column from training set
#train2= train.drop('label', axis = 1, inplace = True)
train2 = train[cols]
# print(dlist)
# print(cols)
print(train2.shape)
print(labels.shape)
```

```python
#reshape and flaten for CNN input
images = train2.values
print(images.shape)
#flatten 42x1
images = np.array([i.flatten() for i in images])
print (images[1].shape)
```

```python
#encode labels
from sklearn.preprocessing import LabelBinarizer
label_binrizer = LabelBinarizer()
labels = label_binrizer.fit_transform(labels) 
labels
# # print(np.unique(unique_value))
# print(np.unique(labels))
```

```python
#split the dataset in to training and test sets
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(images, labels, test_size = 0.2, random_state = 84)
```

```python
#adding bias neuron
x_train = x_train.reshape(x_train.shape[0],x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0],x_train.shape[1], 1)

print(x_train.shape[0])
print(x_train.shape[1])
print(x_train.shape)
```

## CNN Model Building and Training
our main problem is classification/recognition of Ethiopian sign language gesture images to their respective classes using deep learning approaches specifically CNN. The basic steps used in this experiment is described as follows.
* Generate datset from a collection of sign language gesture images
* Load datset from csv and split them for train-test (80:20% ratio)
* Extract class lables from the training set and encode them (LabelBinarizer is used)
* Normalize the landmark values between 0 and 1
* Flatten the input landmark to 1D array
* Build a model architecture (Sequential) with Dense layers
* Train the model and evaluate it and save the model for later use
* Load the trained model and make prediction from signs extacted from live camera feed (Here we used Open CV and MediaPipe to extract landmarks from camera feed)

```python
#CNN model building
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, Dropout
import time

#parameters
batch_size = 128
#J and Z require movement 26-2 =24
num_classes = 7
epochs = 50

model = Sequential()
model.add(Conv1D(64, kernel_size=2, activation = 'relu', input_shape=(42,1) ))
model.add(MaxPooling1D(pool_size = 2))

model.add(Conv1D(64, kernel_size = 3, activation = 'relu'))
model.add(MaxPooling1D(pool_size = 2))

# model.add(Conv1D(64, kernel_size = 3, activation = 'relu'))
# model.add(MaxPooling1D(pool_size = 2))

model.add(Flatten())
model.add(Dense(128, activation = 'relu'))
model.add(Dropout(0.30))

model.add(Dense(num_classes, activation = 'softmax'))

#compile the model
model.compile(loss = keras.losses.categorical_crossentropy, optimizer='adam', metrics=['accuracy'])

#fit
start = time.time()
history = model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs=epochs, batch_size=batch_size)
stop = time.time()
print(f"Training time: {stop - start}s")


#save the model
model.save('CNN_ESL_mediapipe_model.h5')
```

### Plot Accuracy vs Loss graph

```python
print(history.history.keys())
# plot high resolution graph
#plt.figure(figsize=(10, 6),dpi=200)
#  "Accuracy"
plt.figure(figsize=(4, 3),dpi=140)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='center right')
plt.show()
# "Loss"
plt.figure(figsize=(4, 3),dpi=140)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
#Predecting with test images
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred.round())
```

## Testing our trained model using gesture images captured from webcam feed

Here, we use **OpenCV** and **MediaPipe** to generate hand landmarks from live webcam feed. Also we load the previously saved trained CNN model using keras.models.load_model and feed the hand landmarks obtained from mediapipe classifier as an input to the model for prediction.

```python
#imports required modules
import numpy as np
import cv2
import keras
from keras.preprocessing.image import ImageDataGenerator
```

For classifying an input image, first we store all the class labels ofour dataset in a list (list of letters). Then we load our previously trained model using keras and also weget hand landmark coordinates from our camera feed using openCV and MediaPipe. We normalize landmark values, flatten them in to 1D array and feed them to the model for classification. We use np.argmax() to get the index of the nuron with the maximum weight. Finally, we return the recognized letter from the list of leters using the index we got from np.argmax.

```python
#imports
import playsound #audio
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands
# For webcam input:
cap = cv2.VideoCapture(0)

#normalize list function
def normalize(un_list):
    norm_list = [(float(i)-min(un_list))/(max(un_list)-min(un_list)) for i in un_list]
    return norm_list

#letters
leters =['','','','','','','']

#audio files
file_name =['le.WMA','me.WMA','che.WMA','gne.WMA','ke.WMA','ne.WMA','qe.WMA']
        

#load the trained model
model = keras.models.load_model(r"CNN_ESL_mediapipe_model.h5")

with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        success, image = cap.read()
        if not success:
            #skip
            #print("Ignoring empty camera frame.")
            continue

        # Flip the image horizontally for a later selfie-view display, and convert
        # the BGR image to RGB.
        image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
    
        x , y, c = image.shape
        # To improve performance, optionally mark the image as not writeable to
        # pass by reference.
        image.flags.writeable = False
        results = hands.process(image)

        # Draw the hand annotations on the image.
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        landmarks = []
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                for lm in hand_landmarks.landmark:
                    # print(id, lm)
                    lmx = int(lm.x * x)
                    lmy = int(lm.y * y)

                    landmarks.append(lmx)
                    landmarks.append(lmy)
            mp_drawing.draw_landmarks(
            image,
            hand_landmarks,
            mp_hands.HAND_CONNECTIONS,
            mp_drawing_styles.get_default_hand_landmarks_style(),
            mp_drawing_styles.get_default_hand_connections_style())
    
        cv2.imshow('ESL classification using OpenCV, MediaPipe Hands and CNN', image)
    
#         if not landmarks:
#             #print ("List is Empty")
        if landmarks:
            normalized_hand = normalize(landmarks)
            points = np.array(normalized_hand)
            points= points.flatten()
            if points.shape[0] == 42:
            
                x = np.reshape(points,(42,1))
                i = np.reshape(x,(1,x.shape[0],x.shape[1],1))
                
#                 print(points) 
#                 print(i.shape)
                
                #predict image data
                pred = model.predict(i)
                np.argmax(pred)
                print('Pridicted :{}'.format(leters[np.argmax(pred)]))
                
                #play audio of the recognized sign
                playsound.playsound('voice/'+ file_name[np.argmax(pred)], True)
            
                cv2.putText(image, 'Pridicted :'+leters[np.argmax(pred)],(10, 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,0,255), 1)
        # observe the keypress by the user
        keypress = cv2.waitKey(1) & 0xFF

        if keypress == ord("q"):
            break

# free up memory
cap.release()
cv2.destroyAllWindows()
```
